## Scraping http://quotes.toscrape.com with beautifulsoap

#### Fecha de consulta de la pagina web 15/11/2023

```python
# Hacer la solicitud HTTP a la página
url = "http://quotes.toscrape.com"
respuesta = requests.get(url)
respuesta.status_code
```

### el status code 200 es correcto por el cual la pagina nos permite raspar la información.

```python
# Verificar si la solicitud fue exitosa
if respuesta.status_code == 200:
    # Parsear el contenido HTML con Beautiful Soup
    soup = BeautifulSoup(respuesta.text, 'html.parser')

    # Extraer información específica de la página
    for quote in soup.find_all('span', class_='text'):
        print(quote.text)

else:
    print(f"No se pudo acceder a la página. Código de estado: {respuesta.status_code}")
```

### Parseamos el html con beautiful soup con el metodo html.parser

```python
print(soup.find('title').string)
```

```python
autor = soup.find('small', class_='author')
print(autor.text)
```

### imprimimos el el autor mediante la busqueda de class_ "author" 

```python
results = soup.find(class_="container") # ubicar el contenedor pricipal

quotes = results.find_all("div", class_="quote") # ubicar todos los divs que tengas quotes

# listas vacias para almacenar los textos
frases = []
autores = []

# Iterar a través de las citas y extraer la información
for quote in quotes:
    text_element = quote.find("span", class_="text")
    author_element = quote.find("small", class_="author")

    # Añadir los datos a las listas
    frases.append(text_element.text.strip())
    autores.append(author_element.text.strip())

data = {'Quotes': frases, 'Author': autores}
df_frases = pd.DataFrame(data)

df_frases
```

### Al igual que en el caso anterior utilizamos html.parser, identificamos todos los divs y buscamos especificamente por class_ en este caso "mw-body" por ultimo imprimimos el titulo

```python
import bs4 as bs
import requests
from tqdm import tqdm
import pandas

resp = requests.get("https://es.wikipedia.org/wiki/Lionel_Messi")
resp = resp.text
soup = bs.BeautifulSoup(resp, 'lxml') # Esto ayuda a leer el contenido html 

tables = soup.find_all('table', {'class': 'wikitable'})

tables
```

### Realizamos una nueva request en donde nos traemos las etiquetas html, este caso buscamos todas las etiquetas table con class: wikitable que es donde de encuentran las tablas, ya que el continido es muy extendido y confuso reducimos nuestra busqueda a la tabla ubicada en la cuarta ubicacion con las estadisticas de Messi jugando para la seleccion Argentina.

```python
seleccion_argentina = tables[3]
seleccion_argentina
```

### De esta manera podemos ver las etiquetas que corresponden solamente a nuestra tabla en observación.

```python
torneos = []
sedes = []
resultados = []
partidos = []
goles = []
asistencias = []


for row in tqdm(seleccion_argentina.findAll('tr')[1:]):  # Empezamos desde la segunda fila para evitar las etiquetas th
    columns = row.findAll('td')
    if len(columns) >= 6:  # chekear que haya al menos 6 columnas
        torneo = columns[0].text.strip()
        sede = columns[1].text.strip()
        resultado = columns[2].text.strip()
        partido = columns[3].text.strip()
        gol = columns[4].text.strip()
        asistencia = columns[5].text.strip()

        torneos.append(torneo)
        sedes.append(sede)
        resultados.append(resultado)
        partidos.append(partido)
        goles.append(gol)
        asistencias.append(asistencia)
```

#### Creamos lista vacias en donde se almacenaran los datos luego de ser raspados con el bucle, en este caso utilizamos un condicional para indicar que el bucle corra cuando haya al menos 6 columnas, ya que habiamos tenido problemas debido a la longitud de las columnas.
#### en el bucle indicamos que encuentre todos los TR que es donde se encuentra cada celda con la informacion requerida, le pedimos que salte la primera posicion ya que esta incluye las columnnas.
#### Luego cramos las variables para indicar la posicion correcta de la informacion que queremos obtener de cada celda.
#### finalmente con el metodo append agregamos la informacion obtenida mediante el bucle y la asignamos a nuestras listas vacias
#### La libreria tqdm ayuda a visualizar mediante esta barra el progreso de carga de los elementos raspados.

```python
torneos
```

### Comprobamos por ejemplo con la primer lista torneos que los datos han sido raspados correctamente.

```python
df_seleccion_argentina = pd.DataFrame(list(zip(torneos, sedes, resultados, partidos, goles, asistencias)), columns=['Torneo','Sede','Resultado','Partidos','Goles','Asistencias'])
df_seleccion_argentina = df_seleccion_argentina.reset_index(drop=True)
df_seleccion_argentina
```

## Scrapping https://www.filmaffinity.com/es/film685807.html with Beautiful Soap
#### Fecha de consulta de la pagina web 15/11/2023

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
# Hacer la solicitud HTTP a la página
url = "https://www.filmaffinity.com/es/film685807.html"
respuesta = requests.get(url)
respuesta.status_code
```

### Consultado el status code vemos que es correcto

```python
soup = BeautifulSoup(respuesta.text, 'html.parser')
print(soup.find('title').string)
```

### Localizamos el titulo

```python
reviews = soup.find('ul', {'id': 'pro-reviews'})

# Imprimir el contenido del ul seleccionado
print(reviews)
```

### Una vez encontrado el elemento UL  con id = pro-reviews vemos que continene todas las criticas. Dentro de ella las criticas se encuentran en listas ordenadas (LI>) que contienen divs con un interprop = reviewBody para cada review y los autores que se encuentran dentro de un div con class pro-crit-med.

```python
# Crear listas para almacenar datos
autores = []
criticas = []

# Iterar sobre las etiquetas <li> ya que están dentro de las críticas
for critica in reviews.find_all('li'):
    autor_tag = critica.find('div', {'class': 'pro-crit-med'})  # encontrar el autor
    critica_tag = critica.find('div', {'itemprop': 'reviewBody'})  # encontrar cuerpo de la crítica

    # Verificar si tanto autor_tag como critica_tag son diferentes de None
    if autor_tag and critica_tag:
        # Obtener el texto del autor y la crítica
        autor = autor_tag.get_text(strip=True)
        critica = critica_tag.get_text(strip=True)

        # Agregar los datos a las listas
        autores.append(autor)
        criticas.append(critica)

# Crear un DataFrame con las listas
df = pd.DataFrame({'Autor': autores, 'Crítica': criticas})
df
```

### Luego de haber localizado todas las etiquetas procedemos a crear listas vacias para almacenar los datos extraidos, luego realizamos un bucle pasando por todos los li para localizar todos los divs que contienen tanto las criticas como los autores. Alli nos contramos con el problema de que algunas etiquetas no contenian datos de la critica o del autor por lo que nos daba un error el cual fue solucionado con un condicional if indicando que solo substraiga los datos si eran diferentes de None, una vez superado el condicional solo quedo aplicar el metodo append a las listas para luego guardar los datos en un dataframe
